```python
import pandas
import seaborn
import matplotlib.pyplot as mppt

table_advertising = pandas.read_csv('advertising.csv')
# display(table_advertising)

seaborn.pairplot(table_advertising)
mppt.show()
```

```python
import seaborn
import matplotlib.pyplot as mppt

seaborn.heatmap(table_advertising.corr(), annot=True, cmap='Wistia')

mppt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

x = table_advertising.drop('Vendas', axis=1)
y = table_advertising['Vendas']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

linear = LinearRegression()
random_forest = RandomForestRegressor()

linear.fit(x_treino, y_treino)
random_forest.fit(x_treino, y_treino)
```

```python
from sklearn import metrics

linear_test = linear.predict(x_teste)
random_test = random_forest.predict(x_teste)

r2_linear = metrics.r2_score(y_teste, linear_test)
r2_random = metrics.r2_score(y_teste, random_test)

erro_linear = metrics.mean_squared_error(y_teste, linear_test)
erro_random = metrics.mean_squared_error(y_teste, random_test)

print(r2_linear, r2_random)
print(erro_linear, erro_random)
```

```python
 tabela_comparacao = pandas.DataFrame()
 tabela_comparacao['Vendas Reais'] = y_teste
 tabela_comparacao['Previsão'] = random_test 

#  display(tabela_comparacao)
seaborn.lineplot(data=tabela_comparacao)
```
